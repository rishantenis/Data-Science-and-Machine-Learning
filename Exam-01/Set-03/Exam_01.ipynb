```python
# (b) String Manipulation and Functions
def process_string(s):
    s_lower = s.lower()
    vowels = "aeiou"
    vowel_count = 0
    consonant_count = 0

    for char in s_lower:
        if char.isalpha(): 
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    print(f"Number of vowels: {vowel_count}")
    print(f"Number of consonants: {consonant_count}")

    reversed_string = s[::-1]
    return reversed_string

input_string = "Hello World"
print("Revers stri:", process_string(input_string))
```

### Question 2: Control Structures and Data Structures

```python
#(a) Conditional Statements and Loops
numb = list(range(1, 51))


for num in numb:
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
```

```python
# (b) Data Structures Manipulation
students_scores = {
    "Emily": [88, 92, 85],
    "John": [78, 76, 80],
    "Sarah": [90, 91, 89],
    "Michael": [65, 70, 72],
    "Jessica": [95, 98, 94]
}
passed_students = {}

print("Aver Scores:")
for student, scores in students_scores.items():
    average_score = sum(scores) / len(scores)
    print(f"{student}: {average_score:.2f}")

    if average_score >= 85:
        passed_students[student] = average_score
print("\nPassed Students:")
print(passed_students)
```

### Question 4: Data Analysis with Pandas (25 Marks)


```python
# (a) Data Loading and Preprocessing

import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('I:\\SWPU Research\\Research_Task\\BytesOfIntelligence\\Data-Science-and-Machine-Learning\\Exam-01\\Set-03\\Data\\employees.csv')
print("DataFrame Loaded:")
print(df.head())
```

```python
df['JoiningDate'] = pd.to_datetime(df['JoiningDate'])
print("\n'JoiningDate' column converted to datetime:")
print(df.dtypes)
```

```python
today = datetime.now()
df['Experience'] = df['JoiningDate'].apply(lambda date: today.year - date.year - ((today.month, today.day) < (date.month, date.day)))
```

```python
print("\n 'Experience' column added:")
print(df.head())
```

```python
# (b) Data Manipulation and Analysis

filtered_df = df[(df['Department'].isin(['IT', 'HR'])) & (df['Salary'] > 70000)]
print("\nFiltered DataFrame and departments with salary > 70,000):")
print(filtered_df)
```

```python
grouped_df = filtered_df.groupby('Department').agg(
    average_experience=('Experience', 'mean'),
    average_salary=('Salary', 'mean')
).reset_index()

print("\nAverage Experience and Salary:")
print(grouped_df)
```

```python

# (c) Data Exporting
grouped_df.to_csv('department_analysis.csv', index=False)

print("\nexported to 'department_analysis.csv'.")
```
